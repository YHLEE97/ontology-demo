```python
import pandas as pd
from pathlib import Path

# 파일 경로 설정
parquet_path = Path("../data/processed/sgc_energy.parquet")

# Parquet 파일 읽기
if parquet_path.exists():
    df = pd.read_parquet(parquet_path)
    
    # 1. 기본 정보 확인
    print(f"총 페이지 수: {len(df)}")
    print("-" * 50)
    print(df.info())
    print("-" * 50)
    
    # 2. 상위 5개 데이터 확인
    display(df.head())
    
    # 3. 특정 페이지의 마크다운 결과 자세히 보기 (예: 1페이지)
    print("\n[내용 확인 - 1페이지]")
    print(df.iloc[0]['markdown_text']) # 너무 길 수 있으므로 앞부분만 출력
else:
    print(f"파일을 찾을 수 없습니다: {parquet_path}")
```
